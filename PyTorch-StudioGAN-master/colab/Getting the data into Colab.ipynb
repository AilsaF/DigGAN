## Dataset

**Copied from https://github.com/POSTECH-CVLab/PyTorch-StudioGAN/blob/master/README.md**

* CIFAR10: StudioGAN will automatically download the dataset once you execute ``main.py``.

* Tiny Imagenet, Imagenet, or a custom dataset:
  1. download [Tiny Imagenet](https://gist.github.com/moskomule/2e6a9a463f50447beca4e64ab4699ac4) and [Imagenet](http://www.image-net.org). Prepare your own dataset.
  2. make the folder structure of the dataset as follows:

```
┌── docs
├── src
└── data
    └── ILSVRC2012 or TINY_ILSVRC2012 or CUSTOM
        ├── train
        │   ├── cls0
        │   │   ├── train0.png
        │   │   ├── train1.png
        │   │   └── ...
        │   ├── cls1
        │   └── ...
        └── valid
            ├── cls0
            │   ├── valid0.png
            │   ├── valid1.png
            │   └── ...
            ├── cls1
            └── ...
```

## Tiny ImageNet

The fastest way to get Tiny ImageNet dataset in Colab is to follow those steps:
1. download and process the data on your local machine using [tinyimagenet.sh](https://gist.github.com/moskomule/2e6a9a463f50447beca4e64ab4699ac4) as suggested in the [README.md](https://github.com/POSTECH-CVLab/PyTorch-StudioGAN#readme)
2. rename (see structure above)
    * the folder as `TINY_ILSVRC2012` (name used in PyTorch-StudioGAN)
    * `TINY_ILSVRC2012/val/` as`TINY_ILSVRC2012/valid/`
3. compress to zip file (for faster upload)
4. upload to Google Drive into `PyTorch-StudioGAN/data/`
5. unzip with `!unzip <file to unzip>`

Running the [tinyimagenet.sh](https://gist.github.com/moskomule/2e6a9a463f50447beca4e64ab4699ac4) in Colab using `%%shell`, while having your Google Drive mounted, takes  much longer and might even fail.

**Note**, [tinyimagenet.sh](https://gist.github.com/moskomule/2e6a9a463f50447beca4e64ab4699ac4) should be executed as **bash** script. Pasting into **zsh** command line environment **fails** to process the train set properly.
